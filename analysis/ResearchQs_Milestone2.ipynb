## The research questions I will be tackling today is:
### What is the mass distribution of these meteorites? 


```python
import numpy as np
import pandas as pd
import matplotlib.pylab as plt
import seaborn as sns


from scripts import project_functions
df = project_functions.load_and_process(r"C:\Users\Nolan\Desktop\DATA301\course-project-solo_311\data\raw\Meteorite_Landings.xlsx")
```

```python
df['mass'].describe()
```

```python
df.loc[df['mass'] >= 100, 'mass_category'] = '>100g'
df.loc[df['mass'] <= 100, 'mass_category'] = '10g - 100g'
df.loc[df['mass'] <= 10, 'mass_category'] = '<10g' 
```

```python

sns.set_theme(font_scale=1.5,style='ticks')
sns.countplot(data=df,x="mass_category",order=['<10g' ,'10g - 100g','>100g'])
plt.xlabel("Mass Category")
plt.ylabel("Count")
plt.title("Number of Meteorites found by Mass Category")
```

It appears that most of the meteorites discovered have a mass between 10-100g. However it is a fairly even distribution between these categories.

I should investigate to see if the actual mass distribution is similar.

```python
df_masssum = df.groupby('mass_category').sum().reset_index()
```

```python
sns.barplot(data=df_masssum,x='mass_category',y='mass')
plt.xlabel("Mass")
plt.ylabel("Mass Category")
plt.title("   Mass Distribution by Mass Category")
```

*Wow!* It looks like meteorites in the >100g category are the highest contributes to the sum of the masses of all the meteorites together.

To investigate further, I will find the mass distribution of theheaviest meteorites.

```python
df_heaviest = df[['name','mass']].sort_values(by=['mass'],ascending=False)[:20]
```

```python
plt.figure(figsize=(16, 6))
sns.barplot(data=df_heaviest,x='name',y='mass',)
plt.xticks(rotation=45)
plt.xlabel("Name")
plt.ylabel("Mass")
plt.title("Mass Distribution by Mass Category")
```

Looks like Hoba and Cape York have a much greater mass than any other meteorite.

I wonder if I can compare the mass of Hoba and Cape York to the total mass of the rest of the meteorites.

```python
df_mass = df[['name','mass']].sort_values(by=['mass'],ascending=False)
```

```python
df_mass[2:]['mass'].sum()
```

```python
df_mass_dist = df_mass[0:2].reset_index()
```

```python
other = pd.DataFrame([['Rest of Meteorites',df_mass[2:]['mass'].sum()]],columns=['name','mass'])
df_mass_dist=df_mass_dist.append(other)
df_mass_dist=df_mass_dist.drop(columns=['index'])
```

```python
df_mass_dist['mass'] = df_mass_dist['mass'] / df_mass['mass'].sum()
```

```python
plt.figure(figsize=(10, 6))
ax = sns.barplot(data=df_mass_dist,x="name",y='mass')

vals = ax.get_yticks()
ax.set_yticklabels(['{:,.0%}'.format(x) for x in vals])
plt.xlabel("Name")
plt.ylabel("Mass Percentage")
plt.title("Mass Distribution of Meteorites")
```

It appears that the top two asteroids make up 22% of the total mass.
